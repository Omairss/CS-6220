```python
#auto encoder
def auto_encoder(x):
    encoder = tf.layers.dense(x, en_dim, activation=tf.nn.relu)
    out = tf.layers.dense(encoder, out_dim)
    return out
```

```python
def model_fn(features, labels, mode):
    
    x = features['images']
    y = auto_encoder(x)
    
    if mode == tf.estimator.ModeKeys.PREDICT:
        return tf.estimator.EstimatorSpec(mode, predictions=y)
    
    loss_fn = tf.reduce_mean(tf.pow(x - y, 2))
    train_op = tf.train.RMSPropOptimizer(0.01).minimize(loss_fn, global_step=tf.train.get_global_step()) 
    accuracy_ = tf.metrics.accuracy(labels=x, predictions=y)
    
    est_spec = tf.estimator.EstimatorSpec(mode=mode, predictions=y, loss=loss_fn, 
                                          train_op=train_op, eval_metric_ops={'accuracy':accuracy_})
    
    return est_spec
```

### Autoencoder seems have done it's job, but not a good one. Will attempt to classify using kmeans

```python
def performance(labels, clusters):
    
    def get_purity(id1, id2):
        
        id1_unique = np.unique(id1)
        total = 0
        for idx in id1_unique: 
            counts = np.bincount(id2[np.where(id1 == idx)[0]].astype('int'))
            m = np.argmax(np.bincount(id2[np.where(id1 == idx)[0]].astype('int')))
            freq = counts[m] # minlenght in bincounts
            total += freq
        return total / id1.shape[0]
    
    a = get_purity(labels.astype('int'), clusters)
    b = get_purity(clusters, labels)
    return a, b, (2 * a * b) / (a + b)
```

```python
import itertools, operator, scipy
def group_by_label(l):
    it = itertools.groupby(l, operator.itemgetter(1))
    counts = []
    for key, subiter in it:
        counts.append(sum(item[0] for item in subiter))
    return counts

def compute_homogeneity(preds, labels):
    cluster_label_counts = []
    for pred in preds.transpose():
        cluster_label_counts.append(group_by_label([(p,label) for p,label in zip(pred,labels)]))
    
    entropys = []
    for cluster_label_count in cluster_label_counts:
        entropys.append(scipy.stats.entropy(cluster_label_count))
         
    return np.mean(entropys)

def compute_completeness(preds, labels, num_clusters, num_labels):
    label_cluster_counts = {label:np.zeros(num_clusters) for label in range(num_labels)}
    
    for pred, label in zip(preds, labels):
        label_cluster_counts[label] = np.sum([label_cluster_counts[label], pred], axis=0)
    
    entropys = []
    for label_cluster_count in label_cluster_counts.values():
        entropys.append(scipy.stats.entropy(label_cluster_count))
          
    return np.mean(entropys)


def v_measure(preds, labels, num_clusters, num_labels):
    if len(labels) == 0:
        return 1.0, 1.0, 1.0
      
    homogeneity = compute_homogeneity(preds, labels)
    completeness = compute_completeness(preds, labels, num_clusters, num_labels)
    
    if homogeneity==0.0 and completeness==0.0:
        return 0.0, 0.0, 0.0
    v_measure_score = (2.0 * homogeneity * completeness /
                   (homogeneity + completeness))
      
    return homogeneity, completeness, v_measure_score



```

Writing a function to attempt various K and T values

```python
def soft_kmeans(ng_df, T=10, K=10):
    
    data = [row['text'] for index, row in ng_df.iterrows()] 
    vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)
    features = vectorizer.fit_transform(data)
    model = LDA(n_components=T, random_state=666, learning_method='online', n_jobs=-1)
    model.fit(features)
    new_ng_X = model.transform(vectorizer.transform(data))
    GM = GaussianMixture(n_components = K).fit(new_ng_X)
    GM_yhat_prob = GM.predict_proba(new_ng_X)
    return v_measure(GM_yhat_prob, list(ng_df['topic_codes']),K, 20), model
```

```python
from tqdm import tqdm

homogeneity_heatmap = np.zeros([6,4])
completeness_heatmap = np.zeros([6,4])
v_measure_score_heatmap = np.zeros([6,4])

for t, T in tqdm(enumerate(range(4, 20, 2))):
    for k, K in enumerate(range (20, 60, 10)):
        
        h,c,v = soft_kmeans(ng_df, T, K)
        homogeneity_heatmap[t][k] = h
        completeness_heatmap[t][k] = c
        v_measure_score_heatmap[t][k] = v
```
