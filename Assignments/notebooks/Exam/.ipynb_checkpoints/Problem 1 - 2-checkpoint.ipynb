```python
from sklearn.cluster import KMeans
```

```python
def evaluate(labels, clusters):
    
    def purity(id1, id2):
        _id1 = np.unique(id1)
        _sum = 0
        for idx in _id1:
            indices = np.where(id1 == idx)[0]
            _maps = id2[indices].astype('int')
            counts = np.bincount(_maps)
            _max = np.argmax(np.bincount(_maps))
            freq = counts[_max] # minlenght in bincounts
            _sum += freq
        return _sum / id1.shape[0]
    
    n1 = purity(labels.astype('int'), clusters)
    n2 = purity(clusters, labels)
    return n1, n2, (2 * n1 * n2) / (n1 + n2)
```

# ___

```python
from sklearn.feature_selection import chi2
c, p = chi2(X - np.min(X), y)
X_red = X[:, np.argsort(c)[::-1][:50]]
```

```python
km = KMeans(n_clusters=20, n_jobs=-1, random_state=666).fit(X_red)
evaluate(y, km.labels_)
```

# ___

```python
X_out, y_out = X_red[:500], y[:500]
```

```python
np.bincount(y_out.astype('int'))
```

```python
def cheat(labels, clusters, y_out):
    u_labels = np.unique(labels).astype('int')
    cheat_clusters = np.empty(10, dtype='int')
    for label in u_labels:
        indices = np.where(label == labels)[0]
        _maps = clusters[indices].astype('int')
        counts = np.bincount(_maps)
        _max = np.argmax(np.bincount(_maps))
        freq = counts[_max] # minlenght in bincounts
        cheat_clusters[label] = _max
    
    
    for label in u_labels:
        indices = np.where(label == y_out)[0]
        clusters[indices] = cheat_clusters[label]
        
    return clusters
```

```python
labels = cheat(y, km.labels_, y_out)
evaluate(y, km.labels_)
```
