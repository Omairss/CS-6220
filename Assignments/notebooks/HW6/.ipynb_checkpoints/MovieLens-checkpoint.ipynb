```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../data/ml-100k/u.data', delimiter='\t')
data = data.rename(columns = {data.columns[0]: 'user_id', data.columns[1]: 'item_id',
                data.columns[2]: 'rating', data.columns[3]: 'timestamp'})
data = data.set_index('user_id')
```

```python
item_cols = ['movie id',  'movie title', 'release date',  'video release date' ,
              'IMDb URL' , 'unknown' , 'Action', 'Adventure', 'Animation',
              "Children's",  "Comedy",  "Crime", "Documentary", "Drama", "Fantasy",
              "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi",
              "Thriller", "War", "Western"]
```

```python
item = pd.read_csv('../../data/ml-100k/u.item', sep='|', names=item_cols , encoding='latin-1')
item = item.set_index('movie id')
```

```python
user_cols = ['user id', 'age', 'gender', 'occupation', 'zip code']
user = pd.read_csv('../../data/ml-100k/u.user', sep='|', names=user_cols , encoding='latin-1')
user = user.set_index('user id')
```

```python
data.sort_index()
```

```python
data = data.merge(user, how = 'left', left_index = True, right_index = True)
```

```python
data = data.merge(item, right_index = True, left_on = 'item_id', how = 'left')
```

```python
mu_u = data[['rating']].groupby(data.index).sum() / data[['rating']].groupby(data.index).count()
```

```python
sigma_u = (np.power(data[['rating']],2).groupby(data.index).sum() \
           / data[['rating']].groupby(data.index).count()) \
           - np.power(mu_u,2)
```

```python
3 - mu_u.max()
```

```python
#r_mu_sigma = 
data['normalized_rating'] = (data[['rating']] - mu_u)/sigma_u
```

```python
def compute_similarity(data, users):
    
    u = users[0]
    v = users[1]
    
    user_1 = data.loc[data.index == users[0]][['item_id', 'normalized_rating']]
    user_2 = data.loc[data.index == users[1]][['item_id', 'normalized_rating']]
    
    
    rho_uv = user_1.merge(user_2, on='item_id')
    
    return  sum(rho_uv['normalized_rating_x'] * rho_uv['normalized_rating_y'])/len(rho_uv)
```

```python
users = [2,25]
compute_similarity(data, users)
```

```python
def predict(data, u, m, mu_u, sigma_u):
    
    r_vm_rho_uv = 0
    rho_uv = 0
    
    mu_u = float(mu_u.loc[mu_u.index == u, 'rating'])
    sigma_u = float(sigma_u.loc[sigma_u.index == u, 'rating'])
    
    for v in np.unique(data.loc[data['item_id'] == m].index):
        

            r_vm_rho_uv += compute_similarity(data, [u,v]) * float(data.loc[(data.index == v) & (data['item_id'] == m), 'normalized_rating'])
            rho_uv += compute_similarity(data, [u,v])
    
    return mu_u + ((r_vm_rho_uv)/abs(rho_uv) * sigma_u)
```

```python
predict(data, 1, 1, mu_u, sigma_u)
```
